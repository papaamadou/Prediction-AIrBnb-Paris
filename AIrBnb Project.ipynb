```python
data3 = pd.read_csv('ready_listing_airbnb.csv',index_col=0)
```

```python
data3 = data3[(data3.price < 400)&(data3.price > 20)]
```

```python
data3['logprice'] = data3.price.apply(np.log)
```

```python
sns.boxplot(data3.logprice)
```

```python
data3.drop(['price','day'],axis=1,inplace=True)
```

```python
features = [i for i in data3.columns if i!='logprice']
target = 'logprice'
```

```python
X = data3[features]
y = data3[target]
```

Ici je scale les valeurs de X

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=42)
#Regression lineaire
print('LinearRegression()')
lr = LinearRegression()
lr.fit(X_train,y_train)
result1 = lr.predict(X_test) 
print(f'RMSE : {np.sqrt(mean_squared_error(np.exp(result1),np.exp(y_test)))}')
print(f'MAE : {mean_absolute_error(np.exp(result1),np.exp(y_test))}')

print('ElasticNet()')
elas = ElasticNet()
elas.fit(X_train,y_train)
result2 = elas.predict(X_test) 
print(f'RMSE : {np.sqrt(mean_squared_error(np.exp(result2),np.exp(y_test)))}')
print(f'MAE : {mean_absolute_error(np.exp(result2),np.exp(y_test))}')

print('Ridge()')
rd = Ridge()
rd.fit(X_train,y_train)
result3 = rd.predict(X_test) 
print(f'RMSE : {np.sqrt(mean_squared_error(np.exp(result3),np.exp(y_test)))}')
print(f'MAE : {mean_absolute_error(np.exp(result3),np.exp(y_test))}')

print('Lasso()')
ls = Lasso()
ls.fit(X_train,y_train)
result4 = ls.predict(X_test) 
print(f'RMSE : {np.sqrt(mean_squared_error(np.exp(result4),np.exp(y_test)))}')
print(f'MAE : {mean_absolute_error(np.exp(result4),np.exp(y_test))}')


print('KNeighborsRegressor()')
knn = KNeighborsRegressor()
knn.fit(X_train,y_train)
result5 = knn.predict(X_test) 
print(f'RMSE : {np.sqrt(mean_squared_error(np.exp(result5),np.exp(y_test)))}')
print(f'MAE : {mean_absolute_error(np.exp(result5),np.exp(y_test))}')

```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


def rm_stopword(m):
    stop_words = get_stop_words('fr')
    words = word_tokenize(m)
    return [word for word in words if word not in stop_words]
  
```

```python
def rm_stopword_en(m):
    stop_words = get_stop_words('en')
    stop_words = stop_words+['\'','1','2','3','4','5','m2','M2',',']
    words = word_tokenize(m)
    return [word for word in words if word not in stop_words]
  
```
